```python
import matplotlib.pyplot as plt
```

```python
# ChatGPTaccloss_per1000step.csv
# 
steps = [1000, 2000, 3000, 4000, 5000, 6000]
train_loss = [0.7136490925179245, 0.6656241973371826, 0.563629693091901, 0.517809535480507, 0.4372818921381735, 0.37162684041104954]
eval_accuracy = [76.4373546028581, 77.53406447324693, 78.66400797607179, 79.66101694915254, 80.55832502492522, 81.35593220338983]

#  Train Loss
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(steps, train_loss, marker='o', label='Train Loss', color='b')
plt.xlabel('Steps')
plt.ylabel('Loss')
plt.title('Train Loss')
plt.xticks(steps)
plt.grid()
plt.legend()

#  Eval Accuracy
plt.subplot(1, 2, 2)
plt.plot(steps, eval_accuracy, marker='o', label='Eval Accuracy', color='g')
plt.xlabel('Steps')
plt.ylabel('Accuracy(%)')
plt.title('Eval Accuracy')
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```
